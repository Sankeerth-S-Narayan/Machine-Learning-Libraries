```python
import pandas as pd
import numpy as np
```

# join

```python
list_mob = [['vivo',30000],['oppo',40000],['samsung',50000],['apple',70000]]
df_mob = pd.DataFrame(list_mob,columns=['Store','Sales'])
df_mob
```

```python
list_mob1=[['vivo',4],['oppo',3],['samsung',2],['apple',1]]
df_mob1=pd.DataFrame(list_mob1,columns=['Store','Warranty'])
df_mob1
```

```python
df_mob.set_index('Store').join(df_mob1.set_index('Store'),on='Store',how='inner')
```

# merge

```python
pd.merge(df_mob,df_mob1,on="Store")
```

```python
pd.merge(df_mob,df_mob1,on="Store",how="outer")
```

```python
pd.merge(df_mob,df_mob1,on="Store",how="right")
```

# reshaping dataframe

```python
data={ 'Name':["Sam","san","Mia","Roben","John"],
       'Salary':[40000,30000,41000,20000,13000],
       'Gender':['f','m','f','m','m'],
        'Age':[21,24,30,40,35]
     
     
     }
df_emp=pd.DataFrame(data)
df_emp
```

```python
df_reshape=df_emp.melt(id_vars="Gender")
df_reshape
```

```python
df_reshape=df_emp.melt(id_vars="Gender",value_vars="Age")
df_reshape
```

# pivot

```python
pd.pivot_table(df_emp,index=['Gender'],values=['Salary'],aggfunc='sum')
```

# cross tables

```python
data={ 'Car':['bmw','Ford','Honda','Volvo','Ford','bmw'],
      'Sales':[12000,14566,24567,13455,14568,13455],
      'color':['black','black','blue','blue','black','blue']
    
}
df_cars=pd.DataFrame(data)
```

```python
df_cars
```

```python
pd.crosstab(df_cars.Car,df_cars.color,rownames=['Car'],colnames=['color'])
```
