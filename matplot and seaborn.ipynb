```python
import numpy as np
```

```python
import pandas as pd
```

```python
df=pd.read_csv("Automobile.csv")
```

```python
import seaborn as sns
```

```python
sns.relplot(x="width", y="length", data = df)
```

```python
sns.relplot(x="width", y="length",hue = "aspiration", data = df)
```

# catplot 

```python
sns.catplot(x="fuel_type", y="highway_mpg",aspect=3, data=df)
```

```python
#using jitter
```

```python
sns.catplot(x="fuel_type", y="highway_mpg",jitter=False, data=df)
```

```python
#using swarm
```

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="swarm", data=df)
```

```python
sns.catplot(x="body_style", y="highway_mpg",kind="swarm", data=df)
```

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="swarm",hue="aspiration", data=df)
```

# using boxplot

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="box", data=df)
```

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="box",hue="number_of_doors", data=df)
```

# violin plot

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="violin", data=df)
```

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="violin",split=True,data=df)
```

# bar plot

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="bar", data=df)
```

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="point", data=df)
```

```python
sns.catplot(x="fuel_type", y="highway_mpg",kind="swarm",aspect=1,col="number_of_cylinders", data=df)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,diag_kind="kde")
```

```python
sns.countplot('fuel_type',data=df)
```

```python
sns.regplot(x="city_mpg",y="highway_mpg",data=df)
```

```python
sns.heatmap(data=df.corr())
```

```python
sns.heatmap(data=df.corr(),annot=True)
```

```python
sns.heatmap(data=df.corr(),annot=True,cmap="Blues",linewidth=5)
```
